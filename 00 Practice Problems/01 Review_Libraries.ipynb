### Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Sine Wave', color='blue')
plt.title('Sine Wave using Matplotlib')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)                 # Sine of x

# Create a line plot
fig = go.Figure()

# Add a trace (line)
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Sine Wave', line=dict(color='blue')))

# Update layout with title and axis labels
fig.update_layout(
    title='Sine Wave using Plotly',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# Show the plot
fig.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Enable notebook output
output_notebook()

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot
p = figure(title='Sine Wave using Bokeh', x_axis_label='X-axis', y_axis_label='Y-axis')
p.line(x, y, legend_label='Sine Wave', line_width=2, line_color='blue')

show(p)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot
plt.figure(figsize=(10, 5))
sns.lineplot(x=x, y=y, label='Sine Wave', color='blue')
plt.title('Sine Wave using Seaborn')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```
